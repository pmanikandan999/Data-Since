```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("hopeinten.csv")
```

```python
dataset
```

```python
dataset.isnull().sum().fillna(0)
```

```python
dataset['Day'] = pd.to_datetime(dataset['Day'])
```

```python
# Group by date and calculate metrics
result_df = pd.DataFrame(columns=['Date', 'Campaign Name', 'AD Spent', 'No. of Leads', 
                                  'Cost Per Lead', 'CTR', 'Link Click', 'LP Views', 
                                  'Member Passed', 'LP Conversion %'])
```

```python
result_df
```

```python
result_dfs=[]
```

```python
for date, group in dataset.groupby(dataset['Day']):
    ad_spent = group['Amount spent (INR)'].sum()
    num_leads = group['Results'].sum()
    cost_per_lead = ad_spent / num_leads if num_leads != 0 else 0
    ctr = group['CTR (link click-through rate)'].mean()
    link_click = group['Link clicks'].sum()
    lp_views = group['Landing page views'].sum()
    member_passed = (lp_views / link_click) * 100 if link_click != 0 else 0
    lp_conversion = (num_leads / lp_views) * 100 if lp_views != 0 else 0
    
    # Append the calculated metrics for each date to the result DataFrame
    result_group_df = pd.DataFrame({'Date': date.strftime('%d %b (%a)'),
                                  'Campaign Name': 'All Campaigns',
                                  'AD Spent': ad_spent,
                                  'No. of Leads': num_leads,
                                  'Cost Per Lead': cost_per_lead,
                                  'CTR': ctr,
                                  'Link Click': link_click,
                                  'LP Views': lp_views,
                                  'Member Passed': member_passed,
                                  'LP Conversion %': lp_conversion},index=[0])
    result_dfs.append(result_group_df)
    # Concatenate all DataFrames in the list into a single DataFrame
    result_df = pd.concat(result_dfs, ignore_index=True)


# Display or save the result_df as required
    print(result_df)
```

```python
result_df.to_csv('inten_output.csv', index=False)
```
